## Import Libraries and load dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_excel("electric_vehicles_dataset.xlsx")
```

## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Convert data types if needed

```python
# ---------------------- Handling Missing Values ----------------------------------

# Imputation for Autonomous_Level
df['Autonomous_Level'] = df.groupby('Manufacturer')['Autonomous_Level'].transform(
    lambda x: x.fillna(x.mode()[0]))
```

```python
# Renaming Column name
df.rename(columns={'Autonomous_Level': 'Automation_Level'}, inplace=True)
```

```python
# Imputation for Safety_Rating
df['Safety_Rating'] = df.groupby('Manufacturer')['Safety_Rating'].transform(
    lambda x: x.fillna(x.mode()[0]))
```

```python
# Drop CO2_Emissions_g_per_km column, because it's values are full of 0's
df.drop(columns=['CO2_Emissions_g_per_km'], inplace=True)
```

```python
# Checking for duplicates,hence no duplicates are in the dataset.
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
# Changing Datatypes
df['Automation_Level'] = df['Automation_Level'].astype('Int64')
df['Safety_Rating'] = df['Safety_Rating'].astype('Int64')
```

```python
df.dtypes
```

## Stage 3 – EDA and Visualizations

# COUNT PLOT - DISTRIBUTION OF AUTOMATION LEVELS

```python
plt.figure(figsize=(8, 5))
a=sns.countplot(x='Automation_Level', data=df,hue='Automation_Level',palette='viridis',legend=False)
for c in a.containers:
    a.bar_label(c)
plt.title('Distribution of Automation Levels in Electric Vehicles')
plt.xlabel('Automation Level')
plt.ylabel('Number of Vehicles')
plt.show()
```


* The countplot displays the distribution of electric vehicles across different automation levels in the dataset.
The chart shows that Automation Level 5 has the highest number of vehicles compared to other levels.

* This indicates that fully autonomous electric vehicles are more prevalent in this dataset than partially or non-autonomous vehicles.

* Automation_Level used as the categorical variable on the x-axis
Also used as the hue to differentiate each level visually

* The dataset is dominated by Automation Level 5 vehicles, indicating a strong representation of fully autonomous electric vehicles.

## HISTORAM - DISTRIBUTION OF DRIVING RANGE

```python
sns.histplot(df['Range_km'], bins=10, kde=True,color='mediumseagreen',edgecolor="none",alpha=0.7)
plt.title('Distribution of Driving Range of Electric Vehicles')
plt.xlabel('Driving Range (km)')
plt.ylabel('Frequency')
plt.show()
```


* The chart is a histogram showing the distribution of Driving Range of Electric Vehicles (EVs). A histogram shows how frequently different values occur in a dataset

* Range_km – This numerical feature represents the driving range of each EV in kilometers.

* The x-axis - driving range in kilometers.
The y-axis - frequency of how many EVs fall into each range category.The bars indicate the number of EVs within each range interval.
A KDE line (Kernel Density Estimate) is also overlaid to show the probability density—it gives a smooth curve of how the data is distributed.

* This helps understand market trends and what kind of EV ranges are standard versus exceptional.Most EVs have a driving range in a specific cluster (e.g., 250–450 km), showing the most common range among the EVs.
Fewer EVs have very low (<150 km) or very high (>500 km) ranges, suggesting these are less common in the market.

## BOX PLOT - UNITS SOLD 2024

```python
plt.figure(figsize=(6, 6))
sns.boxplot(
    y='Units_Sold_2024',
    data=df,
    boxprops=dict(facecolor='lightblue', edgecolor='black',alpha=0.7),
    medianprops=dict(color='red', linewidth=2),
    whiskerprops=dict(color='green'),
    capprops=dict(color='black'),
    flierprops=dict(marker='o', markerfacecolor='gray', markersize=5)
)
plt.title('Boxplot of Units Sold By 2024')
plt.ylabel('Units Sold on 2024')
plt.show()
```


* The chart shows that most EV models sold around 10,000 units in 2024, with a wide variation in sales and no extreme outliers, generally stable sales performance across models.

* The feature used in this chart is Units_Sold_2024, which represents the number of electric vehicles sold in the year 2024.

* From this plot, showing the overall spread of EV sales in 2024, including the typical sales level, variability across models, and the presence or absence of outliers.

## SCATTER PLOT - BATTERY CAPACITY VS PRICE

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Battery_Capacity_kWh',
    y='Price_USD',
    hue='Battery_Type',
    data=df,palette="Greens")
plt.title('Battery Capacity vs Price by Battery Type')
plt.legend(
    title='Battery Type',
    loc='center left',
    bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```


* This scatter plot shows the relationship between battery capacity and vehicle price, with different colors representing different battery types used in electric vehicles.

* As battery capacity increases, the price of the vehicle generally increases, though prices vary widely depending on the battery type.

* Features are Battery_Capacity_kWh (X-axis), Price_USD (Y-axis), Battery_Type (color / hue)

* It shows how battery size impacts vehicle price and how different battery technologies influence pricing in the EV market.

## BAR PLOT - BATTERY CAPACITY AND BATTERY TYPE

```python
plt.figure(figsize=(8, 6))
sns.barplot(
    x='Battery_Capacity_kWh',
    y='Battery_Type',
    data=df,estimator='mean',hue='Battery_Type',palette='viridis')

plt.title('Average Battery Capacity by Battery Type')
plt.xlabel('Average Battery Capacity (kWh)')
plt.ylabel('Battery Type')
plt.show()
```


* This bar chart shows the average battery capacity for each battery type used in electric vehicles.

* Different battery types have different average capacities, with some battery technologies offering higher energy storage than others.

* Battery_Type (X-axis), Battery_Capacity_kWh (Y-axis – average value)

* It shows how battery capacity varies across battery technologies, helping compare which battery types generally provide larger capacity EVs.

## HEATMAP - RELATIONSHIP ON RANGE, BATTERY CAPACITY, CHARGING TIME

```python
numeric_cols = ['Range_km', 'Battery_Capacity_kWh', 'Charge_Time_hr']
plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='Blues', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Key Electric Vehicle Performance Metrics')
plt.show()
```


* This heatmap shows the correlation between key numerical EV features.

* The chart says  Range, Battery Capacity, and Charging Time have very weak or no correlation with each other, meaning changes in one do not strongly affect the others.

* Range_km, Battery_Capacity_kWh, Charge_Time_hr

* It shows that these EV performance features are largely independent, which helps in understanding feature relationships and avoiding multicollinearity in analysis.

# HEATMAP - TOP MANUFACTURERS - UNITS SOLD BY COUNTRY

```python
grouped = df.groupby(
    ['Country_of_Manufacture', 'Manufacturer']
)['Units_Sold_2024'].sum().reset_index()
top_manufacturers = grouped.groupby('Manufacturer')['Units_Sold_2024'].sum().sort_values(ascending=False) .head(6).index

filtered = grouped[grouped['Manufacturer'].isin(top_manufacturers)]

pivot_table = filtered.pivot(
    index='Country_of_Manufacture',
    columns='Manufacturer',
    values='Units_Sold_2024'
)

plt.figure(figsize=(10,10))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Top 6 Manufacturers – Units Sold by Country (2024)')
plt.show()
```


* This heatmap shows the units sold in 2024 by the top 6 EV manufacturers across different countries.
* Some manufacturers sell more vehicles in specific countries, and the color intensity highlights where each manufacturer performs strongest.
* Country_of_Manufacture(rows), Manufacturer(columns), Units_Sold_2024(values)
* It shows the country-wise sales performance of the top EV manufacturers and identifying strong and weak markets for each manufacturer. Lighter color = lower sales, Darker color = higher sales, Numbers show exact units sold.





## BOXPLOT - BATTERY CAPACITY DISTRIBUTION BY SAFETY RATING & BATTERY TYPE

* This boxplot compares battery capacity across different safety ratings, separated by battery type.
* Battery capacity varies across safety ratings and battery technologies, but there is no single safety rating that consistently has the highest battery capacity.
* Safety_Rating (X-axis), Battery_Capacity_kWh (Y-axis), Battery_Type (color / hue)
* It shows how battery capacity distribution changes based on safety rating and battery type, helping compare capacity spread and outliers across categories.

## AREA PLOT - EV PRICE SEGMENT GROWTH OVER YEARS

```python
colors = ['#4682B4', '#87CEEB', '#708090']
df['Price_Segment'] = pd.cut(df['Price_USD'],bins=[30000,60000,100000,200000],labels=['Mid-Range','Premium','Luxury'])

price_year = df.groupby(['Year', 'Price_Segment']).size().unstack()

price_year.plot(kind='area',figsize=(10,6),alpha=0.7,color=colors)

plt.title('EV Price Segment Growth Over Years')
plt.xlabel('Year')
plt.ylabel('Number of EV Models')
plt.legend(title='Price Segment',loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

* This area chart shows how the number of EV models has changed over the years across three price segments: Mid-Range, Premium, and Luxury.
* The total number of EV models increases over time, with Luxury models consistently dominating, while Mid-Range and Premium segments fluctuate but follow a similar overall trend.
* Year, Price_USD (converted into Price_Segment), and the count of EV models.
* EV market growth is driven mainly by Luxury-priced models, and all price segments contribute steadily to overall EV expansion.

## LINE PLOT - AVG DRIVING RANGE BY BATTERY TYPES

```python
top_batteries = df.groupby('Battery_Type')['Battery_Capacity_kWh'].mean().sort_values(ascending=False).head(3).index.tolist()
df_top_batteries = df[df['Battery_Type'].isin(top_batteries)]
range_year = df_top_batteries.groupby(['Year', 'Battery_Type'])['Range_km'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(
    data=range_year,
    x='Year',
    y='Range_km',
    hue='Battery_Type',
    marker='v',
    palette='Set1')

plt.title('Average Driving Range Trend by Top 3 Battery Types')
plt.xlabel('Year')
plt.ylabel('Average Range (km)')
plt.legend(title='Battery Type', loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

* The chart is a line plot showing the year-wise average driving range of electric vehicles for different battery types.
* From the chart, Nickel-manganese-cobalt batteries appear to be the most commonly used, as they consistently show strong presence and competitive driving range across years.
* The chart uses Year, Battery_Type, and Range_km from the dataset.
* It demonstrates how battery type influences EV performance and how technological advancements increase driving range.

## PAIR PLOT - EV PERFORMACE AND SALES BY PRICE SEGMENT

```python
sns.pairplot(
    df[['Price_USD','Range_km','Safety_Rating','Units_Sold_2024','Price_Segment']],
    hue='Price_Segment',
    diag_kind='kde',palette='viridis')

plt.suptitle('EV Performance and Sales by Price Segment',y=1.02)
plt.show()
```

## JOINT PLOT - AUTOMATION LEVEL AND EV PRICE

```python
sns.jointplot(x='Automation_Level', y='Price_USD', data=df, kind='kde',fill=True,cmap='Blues', height=6)
plt.suptitle('Relationship Between Automation Level and EV Price',y=1.02, fontsize=14)
plt.show()
```

* This chart is a KDE joint plot showing the relationship between electric vehicle automation level and price, along with their individual distributions.
* It shows that EV prices generally increase as the automation level increases, and the darker regions indicate a higher concentration of EVs at those price–automation combinations.
* The features used are Automation_Level and Price_USD.
* This analysis shows that higher automation in EVs is associated with higher pricing, indicating advanced technology leads to higher cost.

## CHOROPLETH (GEOGRAPHICAL) - AVG EV PRICE BY COUNTRY

```python
fig = px.choropleth(
    df.groupby('Country_of_Manufacture')['Price_USD'].mean().reset_index(),
    locations='Country_of_Manufacture',
    locationmode='country names',
    color='Price_USD',
    color_continuous_scale='magma',
    title='Average EV Price by Country')
fig.show()

```

* This is a choropleth world map that shows the average price of electric vehicles for each country using color intensity.
* The features used are Country_of_Manufacture and Price_USD, where price is averaged for each country
* This visualization highlights global differences in EV pricing, helping us compare which countries manufacture higher-priced or more premium electric vehicles.
