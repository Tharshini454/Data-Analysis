```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv"
df = pd.read_csv(url)
df
```

**Exploratory Data Analysis (EDA) and Visualization**

```python
# Campaign Performance:
# Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Acquisition_Cost", y="ROI",color="mediumpurple")
plt.xlabel("Acquisition Cost")
plt.ylabel("ROI")
plt.title("Scatter Plot: Acquisition Cost vs ROI")
plt.grid(True)
plt.show()
```

```python
# Create a bar chart to visualize the average Conversion_Rate for different Channel_Used,categorized by Campaign_Type.
colors = ["steelblue", "mediumseagreen", "indianred", "goldenrod", "mediumpurple"]
grouped = df.groupby(["Channel_Used","Campaign_Type"])["Conversion_Rate"].mean().unstack()
grouped.plot(kind ="bar",figsize=(10,5),color=colors)
plt.title("Average Conversion Rate by Channel and Campaign Type")
plt.xlabel("Channel Used")
plt.ylabel("Average Conversion Rate")
plt.xticks(rotation=0)
plt.legend(title="Campaign Type", bbox_to_anchor=(1.25, 1),loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.
plt.figure(figsize=(10,5))
df.boxplot(column='Engagement_Score', by='Campaign_Type', grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='blue'),
           medianprops=dict(color='red'),
           whiskerprops=dict(color='Purple'),
           capprops=dict(color='Green'))
plt.title('Distribution of Engagement Score by Campaign Type')
plt.suptitle('')  # removes automatic title
plt.xlabel('Campaign Type')
plt.ylabel('Engagement Score')
plt.show()
```

```python
# Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies.
avg_roi = df.groupby('Company')['ROI'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=avg_roi, x='Company', y='ROI',hue="Company", palette='viridis')
plt.title('Average ROI by Company')
plt.xlabel('Company')
plt.ylabel('Average ROI')
plt.show()
```

```python
# Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap.
corr = df[['Engagement_Score', 'Conversion_Rate']].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='Purples')
plt.title('Correlation between Engagement Score and Conversion Rate')
plt.show()
```

```python
# Customer Segmentation:
# Create a count plot to visualize the distribution of Target_Audience.
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Target_Audience',hue="Target_Audience", palette='tab20')
plt.title('Distribution of Target Audience')
plt.xlabel('Target Audience')
plt.ylabel('Count')
plt.show()
```

```python
# Identify which Customer_Segment has the highest Conversion_Rate for each Language using a bar chart.
max_conversion = df.loc[df.groupby('Language')['Conversion_Rate'].idxmax()]
plt.figure(figsize=(10,6))
sns.barplot(data=max_conversion, x='Language', y='Conversion_Rate', hue='Customer_Segment', dodge=False, palette='Set2')
plt.title('Customer Segment with Highest Conversion Rate per Language')
plt.xlabel('Language')
plt.ylabel('Conversion Rate')
plt.legend(title='Customer Segment', bbox_to_anchor=(1.25, 1),loc='upper right')
plt.show()
```

```python
# Visualize the distribution of Acquisition_Cost across each Customer_Segment,categorized by Channel_Used, using a box plot.
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Customer_Segment', y='Acquisition_Cost', hue='Channel_Used', palette='viridis')
plt.title('Acquisition Cost Distribution by Customer Segment and Channel Used')
plt.xlabel('Customer Segment')
plt.ylabel('Acquisition Cost')
plt.legend(title='Channel Used', bbox_to_anchor=(1.25, 1),loc='upper right')
plt.show()
```

```python
# Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages.
avg_conversion = df.groupby('Language')['Conversion_Rate'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=avg_conversion, x='Language', y='Conversion_Rate',hue="Language", palette='cividis')
plt.title('Average Conversion Rate by Language')
plt.xlabel('Language')
plt.ylabel('Average Conversion Rate')
plt.show()
```

```python
# Channel Effectiveness:
# Compare the Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart.
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Channel_Used', y='Engagement_Score', hue='Campaign_Type', palette='PuBu',ci=None)
plt.title('Engagement Score by Channel Used and Campaign Type')
plt.xlabel('Channel Used')
plt.ylabel('Engagement Score')
plt.legend(title='Campaign Type',bbox_to_anchor=(1.25, 1),loc='upper right')
plt.show()
```

```python
# Show the distribution of total ROI across different Channels_Used using a pie chart.
roi_by_channel = df.groupby('Channel_Used')['ROI'].sum()
plt.figure(figsize=(8,5))
colors = sns.color_palette("Pastel1")
plt.pie(roi_by_channel, labels=roi_by_channel.index, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor':'grey'})
plt.title('Distribution of Total ROI by Channel')
plt.show()
```

```python
# Plot a scatter plot to show the relationship between Clicks and Impressions for each Channel_Used.
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Impressions', y='Clicks', hue='Channel_Used', palette='colorblind', s=50)
plt.title('Clicks vs Impressions by Channel')
plt.xlabel('Impressions')
plt.ylabel('Clicks')
plt.legend(title='Channel Used')
plt.show()
```

```python
# Time-Based Analysis:
# Plot the distribution of Duration using a histogram.
plt.figure(figsize=(10,6))
plt.hist(df['Duration'], bins=10, color='mediumslateblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.4)
plt.show()
```

```python
# Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart.
# Convert 'Date' column to datetime, specifying dayfirst=True
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
agg_df = df.groupby(['Date', 'Company'])['Conversion_Rate'].mean().reset_index()
sns.set(style="whitegrid", palette="tab10")
plt.figure(figsize=(14,6))
sns.lineplot(data=agg_df, x='Date', y='Conversion_Rate', hue='Company')
plt.xlabel("Date")
plt.ylabel("Average Conversion Rate")
plt.title("Conversion Rate Over Time for Each Company")
plt.legend(title='Company')
plt.show()
```

```python
agg_engagement = df.groupby('Date')['Engagement_Score'].mean().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(data=agg_engagement, x='Date', y='Engagement_Score', marker='v', color='teal')
plt.xlabel("Date")
plt.ylabel("Average Engagement Score")
plt.title("Trend of Engagement Score Over Time")
plt.show()
```

```python
# Geographic Analysis:
# Determine which location has the highest Acquisition_Cost using a bar chart.
location_acquisition = df.groupby('Location')['Acquisition_Cost'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=location_acquisition.index, y=location_acquisition.values, hue=location_acquisition.index, palette="viridis")
plt.xlabel("Location")
plt.ylabel("Average Acquisition Cost")
plt.title("Average Acquisition Cost by Location")
plt.xticks(rotation=45)
plt.ylim(0, 6000)
plt.show()
```

```python
# Visualize the Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart.
plt.figure(figsize=(12,6))
sns.barplot(data=df,x="Location",y="Conversion_Rate",hue="Target_Audience",palette="Purples",ci=None)
plt.xlabel("Location")
plt.ylabel("Conversion Rate")
plt.title("Conversion Rate by Location Categorized by Target Audience")
plt.xticks(rotation=35)
plt.legend(title="Target Audience", bbox_to_anchor=(1.25, 1),loc='upper right')
plt.show()
```

```python
# Illustrate the proportion of ROI by Location using a pie chart.
roi_location = df.groupby('Location')['ROI'].sum()
plt.figure(figsize=(8,8))
colors = ["#4E79A7", "#F28E2B", "#E15759", "#76B7B2", "#59A14F", "#EDC948"]
plt.pie(roi_location,labels=roi_location.index,autopct='%1.1f%%',pctdistance=0.85,colors=colors)
# Draw circle for donut look
centre_circle = plt.Circle((0,0), 0.60, fc='white')
plt.gca().add_artist(centre_circle)
plt.title("Proportion of ROI by Location")
plt.show()
```
